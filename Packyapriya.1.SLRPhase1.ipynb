```python
#import a library called pandas which has functions to handle database and tables.For internal use, we are storing/naming as pd
import pandas as pd
```

```python
#Data Collection.Storing the CSV file datas to dataset variable. read_csv is a function to read csv file.This function is stored in pandas library.
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#checking if the inputs are stored in memory of dataset variable
dataset
```

```python
#Input Output Split. Saving Input data Years of experience to the input variable-Independent
independent=dataset[["YearsExperience"]]
independent
```

```python
#Input Output Split. Saving Outut data Salary to the output variable-Dependent
dependent=dataset[["Salary"]]
dependent
```

```python
#Train&Test Set Split. using 30% of the input for test set. input and output datas are split into 4 variables. Input has test & train. Output has test and train.
#Hence 4 variable. model selection is a class in sklearn library. train_test_split is a function inside model_selection class
from sklearn.model_selection import train_test_split
```

```python
#Moving the test and train data to respective variables
(X_train,X_test,Y_train,Y_test)=train_test_split(independent,dependent,test_size=0.3,random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
#Model Creation.importing class Linear Regression from linear_model class inside sklearn library.Creating Linear Regression model.The result of that function is stored in 
#regressor variable.fit function is use to plot the train set in the Linear regression model to find the model equation by calculating the weight and bias
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(X_train,Y_train)
```

```python
#w value is stored in coef_variable in the regression function.
weight=regressor.coef_
```

```python
weight
```

```python
#b value is stored in intercept_variable in the regression function.
bias=regressor.intercept_
```

```python
bias
```

```python
#running the model using test data input and predicting the output using the linear regression model we created
Y_pred=regressor.predict(X_test)
```

```python
#Evaluation metrics. using the r2_score function in the metrics class inside the sklearn library.
#If r2 score value in near to 1, it means our mode is working well. if it is close to 0, it is not working well.
from sklearn.metrics import r2_score
```

```python
r_score=r2_score(Y_test,Y_pred)
```

```python
r_score
```

```python
#Save the Model. saving the file under the name mentioned below. wb means write binary.
import pickle
filename="finalized_slr_model_packyapriya.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
#loading the saved model. enter the value inside predict function parameter. run for result.The result shown below is for 15 years of experience
loaded_model=pickle.load(open("finalized_slr_model_packyapriya.sav",'rb'))
```
